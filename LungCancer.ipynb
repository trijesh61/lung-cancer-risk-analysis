# Data Exploration and Visualization

```python
df_cancer=df[df["LUNG_CANCER"]=="YES"]
```

```python
df_cancer.GENDER.value_counts()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(15,10))

# AGE distribution
plt.subplot(2, 2, 1)
sns.histplot(df['AGE'], bins=30, kde=True, color='skyblue')
plt.title('Age Distribution')

# Gender count
plt.subplot(2, 2, 2)
sns.countplot(x='GENDER', hue='GENDER', data=df, palette='pastel', legend=False)
plt.title('Gender Count')

# Smoking count
plt.subplot(2, 2, 3)
sns.countplot(x='SMOKING', hue='SMOKING', data=df, palette='Set2', legend=False)
plt.title('Smoking Status Count')

# Lung Cancer count
plt.subplot(2, 2, 4)
sns.countplot(x='LUNG_CANCER', hue='LUNG_CANCER', data=df, palette='Set1', legend=False)
plt.title('Lung Cancer Distribution')

plt.tight_layout()
plt.show()
```

```python
sns.countplot(x="GENDER" , data=df, hue="LUNG_CANCER", palette="spring")
plt.legend(loc="upper center")
plt.title("Lung Cancer Patients")
plt.show()
```

```python
sns.violinplot(data=df,x="AGE",y="LUNG_CANCER")
plt.title("Lung Cancer patients by Age")
plt.show()
```

```python
g1=df.groupby("SMOKING")
g1.agg(np.size)
```

```python
sns.catplot(data=df,x="SMOKING",hue="LUNG_CANCER",kind="count")
plt.title("Patients who smokes")
plt.show()
```

```python
for i,predictor in enumerate(df.drop(columns=["GENDER","AGE","LUNG_CANCER"])):
    plt.figure(i)
    sns.countplot(x=predictor,data=df,hue="LUNG_CANCER")
```

```python
new_df=df
```

```python
new_df["LUNG_CANCER"]=new_df["LUNG_CANCER"].replace({'YES':1,'NO':0})
new_df.head()
```

```python
plt.figure(figsize=(12, 10))
numeric_df = new_df.drop(columns=["GENDER"])  # Drop non-numeric
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Feature Correlation Heatmap")
plt.tight_layout()
plt.show()
```

## Conclusion

* Females are More suffering From Lung Cancer than Males

* People with **YELLOW FINGERS** has more chance of lung cancer


* People with no peer pressure probably has more Chances of having lung cancer


* People with Coughing  probably has more Chances of having lung cancer
